```python
%matplotlib inline
```

```python
size=10000
shape = 1
```

```python
x1 = np.random.pareto(shape, size)*(np.random.binomial(1,0.5,size)*2-1)
x2 = np.random.pareto(shape, size)*(np.random.binomial(1,0.5,size)*2-1)
x3 = np.random.pareto(shape, size)*(np.random.binomial(1,0.5,size)*2-1)
x4 = np.random.pareto(shape, size)*(np.random.binomial(1,0.5,size)*2-1)
x5 = np.random.pareto(shape, size)*(np.random.binomial(1,0.5,size)*2-1)
```

```python
x1_sum = [x1[:i].sum() for i in range(1,size)]
x2_sum = [x2[:i].sum() for i in range(1,size)]
x3_sum = [x3[:i].sum() for i in range(1,size)]
x4_sum = [x4[:i].sum() for i in range(1,size)]
x5_sum = [x5[:i].sum() for i in range(1,size)]
```

```python
x1_mean = [x1[:i].mean() for i in range(1,size)]
x2_mean = [x2[:i].mean() for i in range(1,size)]
x3_mean = [x3[:i].mean() for i in range(1,size)]
x4_mean = [x4[:i].mean() for i in range(1,size)]
x5_mean = [x5[:i].mean() for i in range(1,size)]
```

```python
p = figure(plot_width=1000, plot_height=500)

# add a line renderer
p.line(range(size), x1, line_width=1)
p.line(range(size), x2, line_width=1, color="red")
p.line(range(size), x3, line_width=1, color="green")
p.line(range(size), x4, line_width=1, color="black")
p.line(range(size), x5, line_width=1, color="blue")
```

```python
show(p)
```

```python
p = plt.figure(figsize=(15, 7))
plt.hist(x1, 1000, density=True)
p.show()
```

```python
ps = figure(plot_width=1000, plot_height=500)

# add a line renderer
ps.line(range(1,size), x1_sum, line_width=1)
ps.line(range(1,size), x2_sum, line_width=1, color="red")
ps.line(range(1,size), x3_sum, line_width=1, color="green")
ps.line(range(1,size), x4_sum, line_width=1, color="black")
ps.line(range(1,size), x5_sum, line_width=1, color="blue")
```

```python
show(ps)
```
